# Constraint Graph

## Up to 4 nodes graph

```python
import networkx as nx
from networkx import isomorphism
four_node_graphs = [g for g in nx.graph_atlas_g() if 1<=len(g.nodes())<=4 and nx.is_connected(g)]
```

```python
for i,g in enumerate(four_node_graphs) : 
    print(f"({i}) nodes : {g.number_of_nodes()}, edges : {g.number_of_edges()}")
```

```python
Constraint = nx.DiGraph()
for i,g in enumerate(four_node_graphs) : 
    Constraint.add_node(i)

for i,G in enumerate(four_node_graphs) : 
    for j,g in enumerate(four_node_graphs) : 
        if j == i : 
            continue
        test = isomorphism.GraphMatcher(G,g).subgraph_is_monomorphic()
        if test : 
            Constraint.add_edge(j,i)
Constraint = nx.transitive_reduction(Constraint)
```

```python
import matplotlib.pyplot as plt 
Constraint = nx.transitive_reduction(Constraint)
for layer, nodes in enumerate(nx.topological_generations(Constraint)):
    # `multipartite_layout` expects the layer as a node attribute, so add the
    # numeric layer value as a node attribute
    for node in nodes:
        Constraint.nodes[node]["layer"] = layer
        Constraint.nodes[node]["order"] = four_node_graphs[node].number_of_nodes()
        Constraint.nodes[node]["size"] = four_node_graphs[node].number_of_edges()
        
# Compute the multipartite_layout using the "layer" node attribute
colors = ['mediumpurple','royalblue','indianred','firebrick', 'orange', 'grey','plum','lightgreen','cornflowerblue','beige','lightcoral','orchid','lightblue','khaki','chocolate','seagreen']
pos = nx.multipartite_layout(Constraint, subset_key="layer")
fig, ax = plt.subplots()
color_map = [colors[node[1]['order']] for node in Constraint.nodes(data=True)]

plt.title("Colored by number of nodes")
nx.draw_networkx(Constraint, pos=pos, ax=ax, node_color=color_map,node_size=200,with_labels=True)
```

```python
fig, axs = plt.subplots(3,3)
for m in range(3) : 
    for n in range(3):
        ax =  axs[m][n]
        ax.tick_params(left = False, right = False , labelleft = False , 
                labelbottom = False, bottom = False) 
        ax.title.set_text(f"{3*m+n}")
        nx.draw_networkx(four_node_graphs[3*m+n],ax=ax,node_size=200)
        
```

```python
Constraint = nx.DiGraph()
for i,g in enumerate(graphs) : 
    Constraint.add_node(i)

for i,G in enumerate(graphs) : 
    for j,g in enumerate(graphs) : 
        if j == i : 
            continue
        test = isomorphism.GraphMatcher(G,g).subgraph_is_monomorphic()
        if test : 
            Constraint.add_edge(j,i)
Constraint = nx.transitive_reduction(Constraint)
```

```python
for k in range(Constraint.number_of_nodes()):
    Constraint.nodes(data=True)[k]["order"] = graphs[k].number_of_nodes()

Constraint.nodes(data=True)
```

```python
for layer, nodes in enumerate(nx.topological_generations(Constraint)):
    # `multipartite_layout` expects the layer as a node attribute, so add the
    # numeric layer value as a node attribute
    for node in nodes:
        Constraint.nodes[node]["layer"] = layer
        
# Compute the multipartite_layout using the "layer" node attribute
pos = nx.multipartite_layout(Constraint, subset_key="layer")
fig, ax = plt.subplots()
colors = ['','blue','red','orange','grey','plum','lightgreen','cornflowerblue','beige','lightcoral']
color_map = [colors[node[1]['order']] for node in Constraint.nodes(data=True)]


nx.draw_networkx(Constraint, pos=pos, ax=ax, node_color=color_map,node_size=30,with_labels=False)
```
